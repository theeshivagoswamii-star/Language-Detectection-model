```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
#load dataset
```

```python
data=pd.read_csv('language.csv')
```

```python
data
```

```python
# Drop null values if any
data.dropna(subset=['Text', 'language'], inplace=True)

X = data['Text'].values
y = data['language'].values
```

```python
# 🔥 FIX 1: .astype(str) lagayein aur .tolist() karke normal python list banayein
# Isse PyArrow ka index internal error bilkul khatam ho jayega
x = data['Text'].astype(str).tolist()
y = data['language'].astype(str).tolist()
```

```python
# 2. Vectorizer apply karein
cv = CountVectorizer()
X = cv.fit_transform(x) # Ab ye bina kisi error ke Matrix bana dega
```

```python
# 🔥 FIX 2: y ko split karte waqt standard numpy array me rakhein
y = np.array(y)
```

```python
# 3. Train Test Split (Ab ye makkhan jaisa chalega)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# 4. Model Train karein
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# 5. Accuracy Check karein
print("Accuracy:", model.score(X_test, y_test))
```

```python
# 6. User Prediction
user = input("Enter the text: ")
user_data = cv.transform([user]).toarray()
output = model.predict(user_data)
print("Predicted Language:", output[0])
```
